```python
import seaborn as sns
tips=sns.load_dataset("tips")
print(tips)
```

```python
tips['day'].value_counts()
```

```python
import pandas as pd
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex'])
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex'],colnames={"sex":"meals"}).rename(columns={"Male":"meals served to meal","Female":"meals served to females"})
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex']).rename(columns={"Male":"meals served by males","Female":"meals served by females"})
```

```python
len(tips.query("sex=='Male' and day=='Sat'"))
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex']).unstack().reset_index()
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex']).unstack().reset_index().rename(columns={0:"meals served"})
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex'],normalize='index')  #for percentage use normalize
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex'],margins=True,margins_name="Total Count") #for total use margins
```

```python
tips.query("sex=='Male' and day=='Thur'")['total_bill'].head()
```

```python
tips.query("sex=='Male' and day=='Thur'")['total_bill'].mean()
```

```python
pd.crosstab(index=tips['day'],columns=tips['sex'],values=tips['total_bill'],aggfunc='mean')   # aggfunc cant be used without values.
```

```python
pd.crosstab(index=tips['day'],columns=[tips['sex'],tips['time']])
```

```python
pd.crosstab(index=tips['day'],columns=[tips['sex'],tips['time']]).rename(columns={"Lunch":"lunch served","Dinner":"dinner served"})
```
