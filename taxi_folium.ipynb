```python
import pandas as pd
import folium
taxi=pd.read_csv('taxi_dataset.csv')
taxi.tail(100)
```

```python
max=50
sf_coordinates=(40.73,-73.98)
map=folium.Map(location=sf_coordinates,zoom_start=12)
for i in taxi[0:max].iterrows():
    folium.Marker(location=[i[1]['pickup_latitude'],i[1]['pickup_longitude']]).add_to(map)
map
```

```python
max=50
sf_coordinate=(40.75,-73.99)
map=folium.Map(location=sf_coordinate,zoom_start=12)
for i in taxi[0:max].iterrows():
    folium.Marker(location=[i[1]['dropoff_latitude'],i[1]['dropoff_longitude']]).add_to(map)
map
```

```python
taxi.dropna(inplace=True)
```

```python
taxi.isnull().sum()
```

```python
taxi.tail()
```

```python
base=folium.Map(location=[taxi["pickup_latitude"].mean(),
              taxi["pickup_longitude"].mean()],
                zoom_start=4,tiles='Stamen Terrain')
station_loc=folium.map.FeatureGroup()

for stn_lat, stn_long in zip(taxi["pickup_latitude"],
                            taxi["pickup_longitude"]):
    station_loc.add_child(
    folium.CircleMarker(
    [stn_lat,stn_long],
    radius=3,
    colors='red',
    fill=True,
    fill_color='red',
    fill_opacity=0.8
    )
    )
base.add_child(station_loc)
```

```python
base=folium.Map(location=[taxi["dropoff_latitude"].mean(),
              taxi["dropoff_longitude"].mean()],
                zoom_start=4,tiles='Stamen Terrain')
station_loc=folium.map.FeatureGroup()

for stn_lat, stn_long in zip(taxi["dropoff_latitude"],
                            taxi["dropoff_longitude"]):
    station_loc.add_child(
    folium.CircleMarker(
    [stn_lat,stn_long],
    radius=5,
    colors='red',
    fill=True,
    fill_color='red',
    fill_opacity=0.8
    )
    )
base.add_child(station_loc)
```
