```python
#crop
image=Image.open('20150816235934.jpg')
```

```python
box=(20,20,20,20)   #(left,upper,right,lower)
```

```python
cropped_image=image.crop(box)
```

```python
cropped_image.save('cropped_image.jpg')
```

```python
#pasting an image into another image
image=Image.open('20150816235934.jpg')
c=Image.open('c.jpg')
image_copy=image.copy()
position=((image_copy.width-c.width),(image_copy.height-c.height))
image_copy.paste(c,position)
```

```python
image_copy.save('pasted_image.jpg')
```

```python
#image rotating
image=Image.open('20150816235934.jpg')
image_rot=image.rotate(90)
image_rot.save('image_rot.jpg')
```

```python
#flipping image transpose
image_flip=image.transpose(Image.FLIP_LEFT_RIGHT)
image_flip.save('image_flip.jpg')
```

```python
image_flip
```

```python
greyscale_image=image.convert('L')
greyscale_image.save('greyscale_image.jpg')
```

```python
from PIL import Image,ImageEnhance
image=Image.open('20150816235934.jpg')
enhancer=ImageEnhance.Sharpness(image)
enhancer.enhance(0).save('image_enhance.jpg')
```
