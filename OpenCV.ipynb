```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('20150816235934.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img_blank=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(img_blank)
```

```python
line_red=cv2.line(img_blank,(0,0),(511,511),(0,255,0),5) #(start)(end)((colour),thickness)
```

```python
plt.imshow(line_red)
```

```python
rectangle=cv2.rectangle(img_blank,(384,0),(510,128),(0,0,255),5)
```

```python
plt.imshow(rectangle)
```

```python
circle=cv2.circle(img_blank,(447,63),63,(0,0,255),-1)  #-1 FILLED CIRCLE
```

```python
plt.imshow(circle)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
text=cv2.putText(img_blank,'OpenCV',(10,500),font,4,(255,255,255),2,cv2.LINE_AA)
plt.imshow(text)
```
