```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import seaborn as sns
```

```python
#importing dataset
data=pd.read_csv('insurance.csv')
data.head()
```

```python
#- Age: Age of primary beneficiary
#- Sex: Primary beneficiary’s gender
#- BMI: Body mass index (providing an understanding of the body, weights that are relatively high or low relative to height)
#- Children: Number of children covered by health insurance / Number of dependents
#- Smoker: Smoking (yes, no)
#- Region: Beneficiary’s residential area in the US (northeast, southeast, southwest, northwest)
#- Charges: Individual medical costs billed by health insurance.
```

```python
data.describe()
```

```python
#getting some insights from visualisation
# Correlation betweeen 'charges' and 'smoker' 
sns.jointplot(x=data['smoker'],y=data['charges'])
```

```python
# we can clearly see that charges of smoker is higher has compared to non-smoker.
```

```python
#Correlation betweeen 'charges' and 'age' 
sns.jointplot(x=data['age'],y=data['charges'])
```

```python
#with increase in the age, cost of medical insurance also increases.
```

```python
#Correlation betweeen 'charges' and 'region' 
sns.jointplot(x=data['region'],y=data['charges'])
```

```python
#the people from southeast have slightly higher charges
```

```python
#Correlation betweeen 'charges' and 'chidlren' 
sns.jointplot(x=data['children'],y=data['charges'])
```

```python
#surprisingly with zero children, it has the highest charges.
```

```python
#Correlation betweeen 'charges' and 'bmi' 
sns.jointplot(x=data['bmi'],y=data['charges'])
```

```python
#it is obvious that people with high bmi will be unhealthy and has higher charges.
```
