```python
import seaborn as sns; sns.set()
flights=sns.load_dataset("flights")
flights.head()
```

```python
ax=sns.countplot(x='passengers',data=flights, palette='pastel')
ax.set_xlabel(xlabel='passengers',fontsize=16)
ax.set_ylabel(ylabel='year',fontsize=16)
ax.set_title(label='passengers',fontsize=16)
```

```python
ax=sns.countplot(x='year',data=flights, palette='pastel')
```

```python
flights=flights.pivot("month","year","passengers")
ax=sns.heatmap(flights)
```

```python
ax=sns.heatmap(flights,annot=True,fmt='d')
```

```python
ax=sns.heatmap(flights,linewidth=.5,annot=True,fmt='d',cmap="YlGnBu")
```

```python
ax=sns.heatmap(flights,cbar=False)
```
