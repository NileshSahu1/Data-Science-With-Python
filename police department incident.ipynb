```python
folium.Marker([37.775420706711, -122.403404791479],tooltip="Weapon Laws",popup='<strong>FIREARM, LOADED, IN VEHICLE, POSSESSION OR USE</strong>',icon=folium.Icon(icon='gun',color='red')).add_to(m)
```
