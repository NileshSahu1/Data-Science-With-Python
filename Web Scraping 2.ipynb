```python
#random articles on wikipedia
import requests
```

```python
#request open url
url="https://en.wikipedia.org/wiki/special:Random"
```

```python
response=requests.get(url)
```

```python
data=response.text
data
```

```python
title_start=data.find("<title>")
print(title_start)
title_end=data.find("</title>")
print(title_end)
```

```python
data[94:137]     #instead of using index number ..use title start end for random wiki content u will get the title
```

```python
print("The title is:",data[title_start+len("<title>"):title_end-len("- Wikipedia")])
```
