```python
#top 10 earn in arsenal replace k and
df['Wage'].replace('€','',inplace=True)
```

```python
df
```

```python
team=df.groupby('Club')
type(team)
arsenal=team.get_group('Arsenal')
arsenal
```

```python
arsenal['Wage']=arsenal['Wage'].str.replace('€','').str.replace('K','')
arsenal
```

```python
arsenal['Wage']=arsenal['Wage'].astype('int')
arsenal
```

```python
earner=arsenal.groupby('Name').max()['Wage']
earner.sort_values(ascending=False)
```

```python
plt.style.use('ggplot')
ars=arsenal.sort_values('Wage',ascending=False).head(10)
```

```python
ax=ars.plot(kind='barh',x='Name',y='Wage',color='#61D199',title='top 10 earners')
ax.set_xlabel("weekly earnings")
```
