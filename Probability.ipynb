```python
#random continous which can measure and discrete which can  count
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Math, Latex   # for latex equations
from IPython.core.display import Image  # for displaying images
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})  #rc for dic format...
```

```python
from scipy.stats import uniform #uniform distributio
```

```python
n=10000
start=10
width=20
data_uniform=uniform.rvs(size=n,loc=start,scale=width)
#rvs(p, loc=0,size=1,random_state=None)Random varities.
```

```python
ax=sns.distplot(data_uniform,
                bins=100,
                kde=True,
                color='skyblue',
                hist_kws={"linewidth":15,'alpha':0.5})    # alpha should be less than one
ax.set(xlabel='Unifrom Distribution',ylabel='Frequency')  # kde kernal density, hist_kws= kernal key words
```
