```python
import seaborn as sns
tips=sns.load_dataset("tips")
print(tips)
```

```python
sns.boxplot(tips["total_bill"])
```

```python
sns.set(style='whitegrid')
sns.boxplot(x=tips["total_bill"],showmeans=True)    #it takes 'x' bydefault
```

```python
tips.describe()   #3 quadraly is 25% 50 and 75 and dot dot is outliers.    outliers=1.5*iqr(iqr=q3-q1)
```

```python
sns.boxplot(x=tips["total_bill"],sym="")    #to remove outliers
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
plt.boxplot(tips["total_bill"])
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x=tips["total_bill"],y=tips["sex"])
```

```python
sns.boxplot(x=tips["total_bill"],y=tips["day"])
```

```python
sns.boxplot(x=tips["total_bill"],y=tips["day"],hue=tips["sex"],palette='Blues') # b capital of blue # and (data=tips)
```

```python
sns.boxplot(x="total_bill",hue="smoker",y="day",data=tips)
```

```python
sns.boxplot(x="total_bill",hue="time",y="day",data=tips)
```

```python
sns.boxplot(x="total_bill",hue="time",y="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",order=["Sun","Sat","Fri","Thur"],data=tips)
```
