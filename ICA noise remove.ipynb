```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import wave
wave1=wave.open("ICA mix 1.wav","r")
wave1
```

```python
signal_1_raw=wave1.readframes(-1)
signal_1=np.fromstring(signal_1_raw,'Int16') #to convert in array
'length:',len(signal_1),'first 100 elements:',signal_1[:100]
```

```python
fs=wave1.getframerate()
timing=np.linspace(0,len(signal_1)/fs,num=len(signal_1))
plt.figure(figsize=(12,2))
plt.title('Recording 1')
plt.plot(timing,signal_1,c='blue')
plt.ylim(-35000,35000)
plt.show()
```

```python
wave2=wave.open("ICA mix 2.wav","r")
wave2
```

```python
signal_2_raw=wave2.readframes(-1)
signal_2=np.fromstring(signal_2_raw,'Int16') #to convert in array
'length:',len(signal_2),'first 100 elements:',signal_2[:100]
```

```python
fs=wave2.getframerate()
timing=np.linspace(0,len(signal_2)/fs,num=len(signal_2))
plt.figure(figsize=(12,2))
plt.title('Recording 2')
plt.plot(timing,signal_2,c='orange')
plt.ylim(-35000,35000)
plt.show()
```

```python
wave3=wave.open("ICA mix 3.wav","r")
wave3
```

```python
signal_3_raw=wave3.readframes(-1)
signal_3=np.fromstring(signal_3_raw,'Int16') #to convert in array
'length:',len(signal_3),'first 100 elements:',signal_3[:100]
```

```python
fs=wave3.getframerate()
timing=np.linspace(0,len(signal_3)/fs,num=len(signal_3))
plt.figure(figsize=(12,2))
plt.title('Recording 3')
plt.plot(timing,signal_3,c='red')
plt.ylim(-35000,35000)
plt.show()
```

```python
X=list(zip(signal_1,signal_2,signal_3))
X[:10]
```

```python
from sklearn.decomposition import FastICA
ica=FastICA(n_components=3)
ica_result=ica.fit_transform(X)
ica_result
```

```python
result_signal_1=ica_result[:,0]
result_signal_2=ica_result[:,1]
result_signal_3=ica_result[:,2]
```

```python
result_signal_2
result_signal_3
result_signal_1
```

```python
plt.figure(figsize=(12,2))
plt.title('Independent Component #1')
plt.plot(result_signal_1,c='red')
plt.ylim(-0.010,0.010)
plt.show()
```

```python
plt.figure(figsize=(12,2))
plt.title('Independent Component #2')
plt.plot(result_signal_2,c='red')
plt.ylim(-0.010,0.010)
plt.show()
```

```python
plt.figure(figsize=(12,2))
plt.title('Independent Component #2')
plt.plot(result_signal_3,c='red')
plt.ylim(-0.010,0.010)
plt.show()
```

```python
from scipy.io import wavfile
# covert to int. map the appropriate range, 
# and icrease the volume a little bit
result_signal_1_int = np.int16(result_signal_1*32767*100)

result_signal_2_int = np.int16(result_signal_2*32767*100)

result_signal_3_int = np.int16(result_signal_3*32767*100)
```

```python
wavfile.write("ICAresult_signal_1.wav", fs, result_signal_1_int)

wavfile.write("ICAresult_signal_2.wav", fs, result_signal_2_int)

wavfile.write("ICAresult_signal_3.wav", fs, result_signal_3_int)
```
