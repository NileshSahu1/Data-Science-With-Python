```python
import requests
#requests opne url
url="https://en.wikipedia.org/wiki/special:random"
#url:uniform resource locator
response=requests.get(url)
```

```python
data=response.text
print(data)
```

```python
#in="title"                        # for true and false
title_start=data.find("<title>")  #find gives index number and if title is not there then ''-1'
print(title_start)
title_end=data.find("</title>")    # end at </title> start at <title>
print(title_end)
```

```python
print("the title is:",data[title_start+len("<title>"):title_end-len("- Wikipedia")])
```
